```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

# Ej 1
Energia en funcion del numero de evoluciones

```python
plt.figure(figsize=(20,10))
datos = pd.read_csv('energia.txt')
plt.plot(datos)
plt.xlabel('# de evoluciones')
plt.ylabel('Energia')
plt.show()
```

# Ej 2
Errores en funcion del numero de evoluciones

```python
plt.figure(figsize=(20,10))
datos = pd.read_csv('errores.txt')
plt.plot(datos)
plt.xlabel('# de evoluciones')
plt.ylabel('# Errores')
plt.show()
```

# Ej 3

```python
plt.figure(figsize=(20,10))
datos = [f for f in os.listdir() if f.startswith('exitos')]
for d in datos:
    datos = pd.read_csv(d)
    M = d.split('_')[1].split('.')[0][1:]
    plt.plot(datos,label=f'M = {M}')
plt.legend()
plt.show()
```

# Ej 4

```python
plt.figure(figsize=(20,10))
datos = pd.read_csv('Resultados/umbrales_promedio/energia.txt')
plt.plot(datos)
plt.xlabel('# de evoluciones')
plt.ylabel('Energia')
plt.show()
plt.figure(figsize=(20,10))
datos = pd.read_csv('Resultados/umbrales_promedio/errores.txt')
plt.plot(datos)
plt.xlabel('# de evoluciones')
plt.ylabel('# Errores')
plt.show()
plt.figure(figsize=(20,10))
datos = [f for f in os.listdir("Resultados/umbrales_promedio/") if f.startswith('exitos')]
for d in datos:
    datos = pd.read_csv("Resultados/umbrales_promedio/"+d)
    M = d.split('_')[1].split('.')[0][1:]
    plt.plot(datos,label=f'M = {M}')
plt.legend()
plt.show()
```
